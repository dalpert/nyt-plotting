```python
import numpy as np
import pandas as pd
import datetime
from time import strftime
from time import gmtime
import matplotlib.pyplot as plt
```

## Data Loading and Manipulation

```python
xword = pd.read_csv('data.csv')
xword.head()
```

```python
# add additional columns
xword['date_obj'] = xword['date'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d').date())
xword['month'] = xword['date_obj'].apply(lambda x: x.month)
xword['date_num'] = xword['date_obj'].apply(lambda x: x.day)
xword['month_name'] = xword['date_obj'].apply(lambda x: x.strftime("%B"))
xword['year'] = xword['date_obj'].apply(lambda x: x.year)
xword['date_label'] = xword['date_obj'].apply(lambda x: x.strftime("%b") + ' ' + x.strftime("%y"))
xword['time_label'] = xword['elapsed_seconds'].apply(lambda x: strftime("%M:%S", gmtime(x)))
```

```python
xword[500:505]
```

```python
# solved only
xword_solved = xword[(xword.solved == 1) & (xword.elapsed_seconds > 0)]
xword_solved.head()
```

```python
# 14d rolling average
xword_rolling = xword_solved.groupby('day')[['date', 'elapsed_seconds']].rolling(14, closed='right').mean().reset_index()
xword_rolling.rename(columns={'level_1':'old_idx'}, inplace=True)
```

```python
xword_rolling.head(20)
```

```python
# join rolling avg and solved dataset to get a table with both dates and rolling averages
rolling_df = xword_rolling.merge(xword_solved[['date', 'date_label', 'time_label', 'date_obj']], left_on='old_idx', right_index=True)
rolling_df.head(20)
```
